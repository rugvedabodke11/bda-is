```python
#Plotting the graph:
fig = plt.figure()
fig.set_size_inches(30,5)
plt.xticks(rotation="vertical")
plt.plot(df["Country/Region"],df["Active"],color="green")
plt.show()
```

```python
#Bar graphs of country/region to active cases:
fig2 = plt.figure()
fig2.set_size_inches(30,5)
country = df["Country/Region"]
active = df["Active"]
plt.xticks(rotation="vertical")
plt.bar(country,active)
plt.show()
```

```python
#Bar graphs of country/region to deaths cases:
fig2 = plt.figure()
fig2.set_size_inches(30,5)
country = df["Country/Region"]
deaths = df["Deaths"]
plt.xticks(rotation="vertical")
plt.bar(country,deaths)
plt.show()
```

```python
#Bar graphs of country/region to recovered cases:
fig3 = plt.figure()
fig3.set_size_inches(30,5)
country = df["Country/Region"]
recovered = df["Recovered"]
plt.xticks(rotation="vertical")
plt.bar(country,recovered)
plt.show()
```
